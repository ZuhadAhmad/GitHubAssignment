```python
#Name: Zuhad Ahmad
#Enrollment no: 01-136242-049

import random

# ----------------------------
# Chatbot Functions
# ----------------------------
def GreetUser():
    print("Welcome, precious... Gollum welcomes you to the Life Simulator!")
    name = input("What’s your name, precious? ")
    age = int(input("How old are you, precious? "))
    user = {"name": name, "age": age}
    print(f"\nGollum: Ahh yesss... {name}, the clever one. We plays, we talks... yes, precious!\n")
    return user

responses = {
    "hi": "Gollum: Hisss... hello, precious!",
    "how are you": "Gollum: Hungry... always hungry. But better now that you’re here.",
    "bye": "Gollum: Nooo! Don’t leaves us, precious! But fine... go then."
}

def GetResponse(user_input):
    return responses.get(user_input.lower(), "Gollum: Hmm... we doesn’t know what that means, precious...")

def EndChat():
    print("Gollum: Goodbye, precious... we will wait in the dark for you... always...")

# ----------------------------
# Game Section
# ----------------------------
stats = {
    "energy": 100,
    "points": 0,
    "mood": "neutral"
}

# Unique Feature: Gollum’s Riddle Challenge
def GollumRiddles():
    riddles = [
        {
            "question": "What has roots that nobody sees, is taller than trees, up, up it goes, and yet never grows?",
            "answer": "mountain"
        },
        {
            "question": "Voiceless it cries, wingless flutters, toothless bites, mouthless mutters. What is it?",
            "answer": "wind"
        },
        {
            "question": "This thing all things devours: birds, beasts, trees, flowers; gnaws iron, bites steel; grinds hard stones to meal; slays king, ruins town, and beats high mountain down.",
            "answer": "time"
        },
        {
            "question": "Alive without breath, as cold as death; never thirsty, ever drinking, all in mail never clinking.",
            "answer": "fish"
        },
        {
            "question": "It cannot be seen, cannot be felt, cannot be heard, cannot be smelt. It lies behind stars and under hills, and empty holes it fills.",
            "answer": "dark"
        }
    ]
    print("\nGollum: Yesss... let's play riddles, precious. Answer if you can, or lose your points, yesss...")

    score = 0
    for riddle in riddles:
        print("\nRiddle:", riddle["question"])
        guess = input("Your answer: ").lower().strip()
        if guess == riddle["answer"]:
            print("Gollum: Clever, precious! Very clever! You gets one point!")
            score += 1
        else:
            print(f"Gollum: WRONG! The answer was '{riddle['answer']}', stupid hobbitses!")
            score -= 1

    print(f"\nYour riddle score: {score}")
    if score > 0:
        print("Gollum: Yesss... we likes you now, clever precious!")
    else:
        print("Gollum: Nooo, stupid! You loses! My precioussss wins!")
    return score

def ShowStatus():
    print("\n Current Stats ")
    print(f"Energy: {stats['energy']}")
    print(f"Points: {stats['points']}")
    print(f"Mood: {stats['mood']}")
    print("----------------------------")

def UpdateStats(activity):
    if activity == "study":
        stats["energy"] -= 20
        stats["points"] += 10
        stats["mood"] = "tired"
        print("You studied hard, precious. Gollum is impressed but sleepy now.")
    elif activity == "eat":
        stats["energy"] += 15
        stats["mood"] = "happy"
        print("Yum! We eats and feels strong again, precious!")
    elif activity == "play":
        stats["energy"] -= 10
        stats["points"] += 5
        stats["mood"] = "happy"
        print("We plays games, precious! Fun, fun, yes!")
    elif activity == "sleep":
        stats["energy"] = 100
        stats["mood"] = "neutral"
        print("Gollum naps... sleeps like a rock, precious.")
    else:
        print("Gollum: What nonsense is this, precious? That’s not an activity!")

    if stats["energy"] > 100:
        stats["energy"] = 100
    if stats["energy"] < 0:
        stats["energy"] = 0

    ShowStatus()

def EndGame(user):
    print("\nGame Over, precious!")
    print(f"{user['name']}, you earned {stats['points']} points!")
    print(f"Energy left: {stats['energy']}")
    print(f"Mood: {stats['mood']}")
    print("Gollum: We likes you, precious... maybe. Until next time...")

def GameMenu(user):
    print("\n Welcome to the Gollum Life Simulator ")
    while True:
        print("\nChoose an activity:")
        print("1. Study")
        print("2. Eat")
        print("3. Play Games")
        print("4. Sleep")
        print("5. Play Riddles with Gollum (Unique Feature)")
        print("6. Quit")

        choice = input(" Your choice: ").lower()
        if choice in ["1", "study"]:
            UpdateStats("study")
        elif choice in ["2", "eat"]:
            UpdateStats("eat")
        elif choice in ["3", "play"]:
            UpdateStats("play")
        elif choice in ["4", "sleep"]:
            UpdateStats("sleep")
        elif choice in ["5", "riddle", "riddles"]:
            score = GollumRiddles()
            stats["points"] += score * 5
        elif choice in ["6", "quit"]:
            EndGame(user)
            break
        else:
            print("Gollum: We doesn’t understand that, precious... try again!")

# ----------------------------
# Main Chat Loop
# ----------------------------
def main():
    user = GreetUser()
    while True:
        user_input = input(" You: ").lower()
        if "play" in user_input or "game" in user_input:
            ans = input("Gollum: Play Life Simulator with Gollum, yesss? (yes/no) ").lower()
            if ans == "yes":
                GameMenu(user)
            else:
                print("Gollum: Fine, precious. We waits in the shadows then...")
        elif user_input == "bye":
            EndChat()
            break
        else:
            print(GetResponse(user_input))

# ----------------------------
# Run the Program
# ----------------------------
if __name__ == "__main__":
    main()
```
